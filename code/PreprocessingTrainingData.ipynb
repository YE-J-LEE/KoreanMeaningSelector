```python
import re, openpyxl, nltk
from khaiii import KhaiiiApi
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('tagsets')
api = KhaiiiApi()

def tokenizeEN(sentence, lang="en"):
  morphs = []
  tokens = nltk.word_tokenize(sentence)
  for tag in nltk.pos_tag(tokens):
      morphs.append(tag[0].lower() + ":" + tag[1] + '_' + lang)

  return ' '.join(morphs)


def tokenizeKO(sentence, lang="ko"):
  morphs = []
  for word in api.analyze(sentence):
      for i in range(len(word.morphs)):
            morphs.append(str(word.morphs[i]).replace("/", ":") + '_' + lang)

  return ' '.join(morphs)

print(tokenizeEN("He's..?! very, angry.?"))
print(tokenizeKO("'6, 5, 7, 8' 사이즈가 몇 개나 더 재 입고 될지 알려주세요."))
```

```python
def readAndWrite(src, dst, startRow, col1, col2):
  wb = openpyxl.load_workbook(filename=src)
  ws = wb.active
  f = open(dst, 'a')
  first = True
  test = ''

  for row in ws.iter_rows(min_row=startRow, max_row=ws.max_row):
    data = tokenizeKO(row[col1].value) + " <<split>> " + tokenizeEN(row[col2].value) + '\n'
    f.write(data)
    if first:
      first = False
      test = data
 
  f.close()
  wb.close()
  return test
```
